➡ Outlier가 너무 많아서, 가격 분포 그래프를 그리기 어려움이 있음

> PRICE의 Outlier 제거 (IQR 기반)

````python
def remove_outliers(df, column):
  q1 = df[column].quantile(0.25)
  q3 = df[column].quantile(0.75)
  iqr = q3 - q1
  boundary = 1.5*iqr

  return df[(df[column] < q3+boundary) & (df[column] > q1-boundary)]
````

````python
df1_sales_no_outliers = remove_outliers(df1_sales, 'PRICE')
````

````python
sns.boxplot(y='PRICE_M', data=df1_sales_no_outliers)
````

### 지역별 판매 가격 시각화

````python
df_location_sales['PRICE_M'].describe()
````

````python
q1 = df_location_sales['PRICE_M'].quantile(0.25)
q3 = df_location_sales['PRICE_M'].quantile(0.75)
iqr = q3 - q1
boundary = 1.5*iqr

def get_sales_price_range(price):
  if price <= q1 - boundary:
    return 'very low'
  elif price <= q1:
    return 'low'
  elif price <= q3:
    return 'medium'
  elif price <= q3 + boundary:
    return 'high'
  elif price > q3 + boundary:
    return 'very high'
````

````python
df_location_sales['PRICE_M_Range'] = df_location_sales['PRICE_M'].apply(get_sales_price_range)
````

````python
plt.style.use('dark_background')

sns.scatterplot(x='LONGITUDE', y='LATITUDE', hue='PRICE_M_Range', data = df_location_sales, alpha=0.5, s= 10)
plt.title('Sales Price by Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid()
plt.show()
````

➡ Latitude(위도)가 낮을 수록 판매 가격이 낮은 경향을 보임

➡ Latitude(위도)가 높을 수록 판매 가격이 높은 경향을 보임

## 주택 분석 결과 요약

```
- 주택 건축 :
  1960-70년대, 2000년대에 건축이 활발했으며, 최근에는 Condominium과 Residential 건축이 비슷한 비율로 진행되고 있습니다.

  2층 주택이 가장 선호되며, 연립 주택 중심부, 단독주택, 반독립주택 순으로 선호됩니다.

- 주택 가격 :
  가격은 $0.1~0.3M 사이의 주택이 가장 많으며, 가격이 상승할수록 주택 수는 감소하는 경향을 보입니다.

- 위치 :
  1901-1930년대 건축물은 도시 중심부에 밀집되어 있고, 1991년 이후 건축물은 도시 전역에 분포되어 있습니다.
  외곽 지역은 Latitude가 높은 지역에서 판매가 활발하며, 가격도 높은 경향을 보입니다.

- 인구 및 소득 :
  북서부 : 백인이 다수 거주, 높은 소득 수준, 주택 가격도 높은 경향을 띕니다.
  남동부 : 흑인이 다수 거주, 낮은 소득 수준, 주택 가격 저렴한 편입니다.
  도시 중심부 : 다양한 인종이 밀집, 아시아인이 상대적으로 높은 비중을 차지합니다.

- 상권 활성화 요소 :
  도시 중심부 : 높은 인구 밀도와 다양한 인종 구성으로 다양한 소비 패턴이 발생할 수 있습니다.
  북서부 : 고소득층과 주택 가격이 높은 지역으로 프리미엄 전략이 적합합니다.
  남동부 : 저소득층이 집중된 지역으로 가격 경쟁력을 활용한 접근이 필요합니다.
```



## 주택 관련 전략

```
(1) 도시 외곽, 특히 Latitude가 높은 지역 :
  2층 주택, 단독주택, 반독립주택 중심의 가족 단위 고객을 타겟팅.
  특히, 가족 단위로 거주하는 지역에는 공원, 학교와 같은 부대 시설이 Alpha의 판매 목적과 부합할 수 있는지 고려해볼 필요가 있음.

(2) 도시 중심부 :
  리모델링이 필요한 고급 주택 거주자나 복합 주택 거주자를 타겟팅.
  또한 역사적 가치가 있는 건물의 경우 관광지로 이용될 가능성이 있으므로, 여행객을 타겟으로 하는 경우도 고려해야 함.
```

## 상권 개발 전략

```
(1) 도시 중심부 (다인종 거주 지역) :
인구 밀집도와 다양한 소비 계층을 활용해 다목적 매장(식료품, 생활용품, F&B 등) 입점.

(2) 북서부 (고소득층 거주 지역) :
프리미엄 제품 라인을 주력으로 한 고급 매장 런칭.
주택 가격이 높은 지역으로, 인테리어 및 서비스 의 차별화된 품질에 투자하는 방안을 추천함.

(3) 남동부 (저소득층 거주 지역) :
저렴한 가격의 필수품 및 대량 구매가 가능한 대형 할인 매장 전략.
지역 커뮤니티 지원 프로그램(장학금, 직업 훈련 등)을 통해 긍정적인 브랜드 이미지 확보 가능.

```

```
<지역별로 특화된 상품 판매하기>

```

- 고소득 및 저소득 지역에서 각각의 차별화된 전략을 통해 수익성을 극대화함.
- 도시 중심부 소비자들의 다양한 니즈를 반영하는 트렌디한 브랜드 이미지를 구축하고, 브랜드 인지도를 높일 수 있음.
