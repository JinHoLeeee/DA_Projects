> PRICEì˜ Outlier ì œê±° (IQR ê¸°ë°˜)

````python
def remove_outliers(df, column):
  q1 = df[column].quantile(0.25)
  q3 = df[column].quantile(0.75)
  iqr = q3 - q1
  boundary = 1.5*iqr

  return df[(df[column] < q3+boundary) & (df[column] > q1-boundary)]
````

````python
df1_sales_no_outliers = remove_outliers(df1_sales, 'PRICE')
````
